```python
#Record of Presidential Margin: 2012

#clean data

#read in county-level data
pres_res_12 = pd.read_csv('2012_pres_result.csv')
#reformat
pres_res_12 = pres_res_12.replace(',','', regex=True)

pres_res_12['Rep_Total'] = pres_res_12['Rep_Total'].astype(int)

pres_res_12['Dem_Total'] = pres_res_12['Dem_Total'].astype(int)

#Create column of total 2 party vote
pres_res_12['2_party_vote'] = pres_res_12['Dem_Total'] + pres_res_12['Rep_Total']
#Create Rep and Dem percentage columns
pres_res_12['Rep_pct'] = pres_res_12['Rep_Total'] / pres_res_12['2_party_vote']

pres_res_12['Dem_pct'] = pres_res_12['Dem_Total'] / pres_res_12['2_party_vote']

#Create Rep Margin column
pres_res_12['Rep_Margin'] = pres_res_12['Rep_pct'] - pres_res_12['Dem_pct']

#rename county column
pres_res_12 = pres_res_12.rename(columns={"Unnamed: 0": "County"}, errors="raise")

#Only include county and marhin column
pres_res_12 = pres_res_12[['County', 'Rep_Margin']]

#remove nas
pres_res_12 = pres_res_12.replace([np.inf, -np.inf], np.nan).dropna(axis=1)



```

```python
#Record of Presidential Margin: 2008

#clean data

#read in county-level data
pres_res_08 = pd.read_csv('2008_pres_result.csv')

#format
pres_res_08 = pres_res_08.replace(',','', regex=True)

pres_res_08['Rep_total'] = pres_res_08['Rep_total'].astype(int)

pres_res_08['Dem_total'] = pres_res_08['Dem_total'].astype(int)

#Create column of total 2 party vote
pres_res_08['2_party_vote'] = pres_res_08['Dem_total'] + pres_res_08['Rep_total']

#Create Dem and Rep percentage columns
pres_res_08['Rep_pct'] = pres_res_08['Rep_total'] / pres_res_08['2_party_vote']

pres_res_08['Dem_pct'] = pres_res_08['Dem_total'] / pres_res_08['2_party_vote']

#Create Republican Margin column

pres_res_08['Rep_Margin'] = pres_res_08['Rep_pct'] - pres_res_08['Dem_pct']

#Only include county and margin columns
pres_res_08 = pres_res_08[['County', 'Rep_Margin']]

pres_res_08 = pres_res_08.replace([np.inf, -np.inf], np.nan).dropna(axis=1)

```
